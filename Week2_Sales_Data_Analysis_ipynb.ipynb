```python
import pandas as pd

# Load the sales analysis output file
df = pd.read_csv("week2_sales_analysis_output.csv")

# Show first 5 rows
df.head()
```

```python
# Basic information about the dataset
df.info()
```

```python
# Calculate total revenue using numeric columns only
total_revenue = df.select_dtypes(include='number').sum().sum()

# Total number of orders
total_orders = len(df)

print("Total Revenue:", total_revenue)
print("Total Orders:", total_orders)

```

```python
# Re-export the file (safe step)
df.to_csv("week2_sales_analysis_output.csv", index=False)

print("Output CSV ready")
```
